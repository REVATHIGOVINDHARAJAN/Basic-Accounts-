```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the table
data = {
    "Type of Account": ["Real Account", "Personal Account", "Nominal Account"],
    "Golden Rules": [
        "🔹 **Debit** what comes into the business\n🔹 **Credit** what goes out from the business",
        "🔹 **Debit** the receiver\n🔹 **Credit** the giver",
        "🔹 **Debit** the expense or loss of the business\n🔹 **Credit** the income or gain of the business"
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis with a larger size
fig, ax = plt.subplots(figsize=(8, 4))

# Set Title with Color
title_text = "Golden Rules of Accounting"
ax.set_title(title_text, fontsize=14, fontweight='bold', color='#D35400', pad=15, loc='center')

# Hide axes
ax.xaxis.set_visible(False) 
ax.yaxis.set_visible(False)
ax.set_frame_on(False)

# Create the table with Markdown formatting
table = ax.table(
    cellText=df.values, 
    colLabels=df.columns, 
    cellLoc='left', 
    loc='center', 
    bbox=[0, 0, 1, 1]
)

# Adjust column widths
table.auto_set_font_size(False)
table.set_fontsize(10)

# Set column widths manually for better alignment
table.auto_set_column_width([0, 1])  
table.scale(1.4, 1.4)  

# Proper cell formatting with colors and alignment
for (i, j), cell in table.get_celld().items():
    if i == 0:  # Header row
        cell.set_facecolor('#3498DB')  # Blue background
        cell.set_fontsize(11)
        cell.set_text_props(weight='bold', ha='center', color='white')  # White text
        cell.set_edgecolor('black')
        cell.set_linewidth(1.5)  # Thicker header border
    else:  
        if j == 1:  # Golden Rules Column
            cell.set_text_props(ha='left', va='center', fontsize=10, fontweight='normal', linespacing=1.2)
        else:  # Type of Account Column
            cell.set_text_props(ha='center', va='center', fontsize=10, fontweight='bold', color='#2C3E50')  # Dark blue text
            
        cell.set_facecolor('#ECF0F1')  # Light grey background for readability
        cell.set_edgecolor('black')
        cell.set_linewidth(1)

# Adjust layout to fit the table properly
plt.subplots_adjust(left=0.1, right=0.9, top=0.8, bottom=0.1)

# Show the table
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define categories and subcategories
categories = {
    "ACCOUNTS": ["PERSONAL ACCOUNTS", "IMPERSONAL ACCOUNTS"],
    "PERSONAL ACCOUNTS": ["Proprietor", "Creditors", "Debtors"],
    "IMPERSONAL ACCOUNTS": ["REAL ACCOUNTS", "NOMINAL ACCOUNTS"],
    "REAL ACCOUNTS": ["Assets (Meant for Use)", "Goods (Meant for Resale)"],
    "NOMINAL ACCOUNTS": ["Expenses & Losses", "Incomes & Gains"]
}

# Add edges based on dictionary structure
for parent, children in categories.items():
    for child in children:
        G.add_edge(parent, child)

# Create a hierarchical layout
pos = nx.spring_layout(G, seed=42, k=0.8, scale=2)

# Assign colors
color_map = {
    "ACCOUNTS": "red",
    "PERSONAL ACCOUNTS": "blue",
    "IMPERSONAL ACCOUNTS": "blue",
    "REAL ACCOUNTS": "green",
    "NOMINAL ACCOUNTS": "green"
}
node_colors = [color_map.get(node, "purple") for node in G.nodes]

# Draw the graph
plt.figure(figsize=(12, 7))
nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color="gray",
        node_size=3500, font_size=10, font_weight="bold", linewidths=1.5)

# Title for the mind map
plt.title("Mind Map: Types of Accounts", fontsize=14, fontweight="bold", color="darkblue")
plt.show()
```

 ####  3.Prepare Purchase  Returns Book and Sale  returns  Book From the following  data:
 ####   1987 Aug 1   Purchased Goods Returned to Senthil             Rs.205
 ####   1987 Aug 3   Received  Goods Returned By  Natarajan          Rs.300
 ####   1987 Aug 5   Goods Returned  to Kannan                       Rs.500
 ####   1987 Aug 7   Sales Returns of Rs.1,260 By Mathavan 
 ####   1987 Aug 15  Returned  Defective goods returned by Rajan     Rs.1,280
 ####   1987 Aug 18  Damaged  Goods returned by  Murali              Rs.1,120
 ####   1987 Aug 23  Ontward Returns to Kanagasabai                  Rs.  275
 ####   1987 Aug 29  Inward Returns By Swaminathan                   Rs.  750
 ####   1987 Aug 30  Returned  inferior Goods to Sankar              Rs.  890
 ####   1987 Aug 31  Selvan  Returned  Goods to us                   Rs.1,330

```python
import matplotlib.pyplot as plt
import pandas as pd

# Purchase Returns Data
purchase_returns = [
    ["1987-08-01", "Senthil", "PN001", "205"],
    ["1987-08-05", "Kannan", "PN002", "500"],
    ["1987-08-23", "Kanagasabai", "PN003", "275"],
    ["1987-08-30", "Sankar", "PN004", "890"]
]

# Sales Returns Data
sales_returns = [
    ["1987-08-03", "Natarajan", "CN001", "300"],
    ["1987-08-07", "Mathavan", "CN002", "1,260"],
    ["1987-08-15", "Rajan", "CN003", "1,280"],
    ["1987-08-18", "Murali", "CN004", "1,120"],
    ["1987-08-29", "Swaminathan", "CN005", "750"],
    ["1987-08-31", "Selvan", "CN006", "1,330"]
]

# Convert Data to DataFrame
purchase_df = pd.DataFrame(purchase_returns, columns=["Date", "Supplier Name", "Debit Note No", "Amount (Rs.)"])
sales_df = pd.DataFrame(sales_returns, columns=["Date", "Customer Name", "Credit Note No", "Amount (Rs.)"])

# Compute Totals
purchase_total = sum(int(row[3].replace(",", "")) for row in purchase_returns)
sales_total = sum(int(row[3].replace(",", "")) for row in sales_returns)

# Append Total Row
purchase_df.loc[len(purchase_df)] = ["", "TOTAL", "", f"{purchase_total}"]
sales_df.loc[len(sales_df)] = ["", "TOTAL", "", f"{sales_total}"]

# Create Matplotlib Figure
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
fig.suptitle("Purchase Returns & Sales Returns Book", fontsize=16, fontweight='bold', color="#3498DB", y=1.05)

# Function to Create Colorful Ledger Table
def create_ledger_table(ax, df, title, header_color, total_color):
    ax.set_title(title, fontsize=14, fontweight='bold', color=header_color, pad=10)
    ax.axis('tight')
    ax.axis('off')
    
    # Create the Table
    table = ax.table(
        cellText=df.values, 
        colLabels=df.columns, 
        cellLoc='center', 
        loc='center', 
        bbox=[0, 0, 1, 1]
    )
    
    # Adjust Font Size & Column Width
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.auto_set_column_width([0, 1, 2, 3])
    table.scale(1.2, 1.2)
    
    # Apply Colors & Formatting
    for (i, j), cell in table.get_celld().items():
        if i == 0:  # Header row
            cell.set_facecolor(header_color)
            cell.set_text_props(weight='bold', color='white', ha='center')
            cell.set_edgecolor('black')
        elif i == len(df) - 1:  # Total row
            cell.set_facecolor(total_color)
            cell.set_text_props(weight='bold', color='white', ha='center')
        else:
            cell.set_text_props(ha='center', fontsize=10, color='black')

# Generate Purchase Returns Table
create_ledger_table(axes[0], purchase_df, "PURCHASE RETURNS ACCOUNT", "#2980B9", "#E74C3C")

# Generate Sales Returns Table
create_ledger_table(axes[1], sales_df, " SALES RETURNS ACCOUNT", "#27AE60", "#E67E22")

# Show the Matplotlib Tables
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Trial Balance Data
trial_balance_data = [
    [1, "Capital", "", "", "9,000"],
    [2, "Plant and Machinery", "", "12,000", ""],
    [3, "Purchase", "", "8,000", ""],
    [4, "Sales", "", "", "12,000"],
    [5, "Sundry Creditors", "", "", "8,000"],
    [6, "Bank Loan", "", "", "22,000"],
    [7, "Rent Outstanding", "", "", "1,000"],
    [8, "Opening Stock", "", "2,000", ""],
    [9, "Sales Returns", "", "4,000", ""],
    [10, "Investments", "", "14,000", ""],
    [11, "Debtors", "", "12,000", ""]
]

# Convert Data to DataFrame
trial_balance_df = pd.DataFrame(trial_balance_data, columns=["S.NO", "Name of Account", "LF", "Debit Balance (Rs.)", "Credit Balance (Rs.)"])

# Compute Totals
debit_total = sum(int(row[3].replace(",", "")) if row[3] else 0 for row in trial_balance_data)
credit_total = sum(int(row[4].replace(",", "")) if row[4] else 0 for row in trial_balance_data)

# Append Total Row
trial_balance_df.loc[len(trial_balance_df)] = ["", "[TOTAL", "", f"{debit_total}", f"{credit_total}"]

# Create Matplotlib Figure
fig, ax = plt.subplots(figsize=(10, 6))
fig.suptitle("TRIAL BALANCE", fontsize=16, fontweight='bold', color="#3498DB", y=1.05)

# Create the Table
table = ax.table(
    cellText=trial_balance_df.values, 
    colLabels=trial_balance_df.columns, 
    cellLoc='center', 
    loc='center', 
    bbox=[0, 0, 1, 1]
)

# Adjust Font Size & Column Width
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width([0, 1, 2, 3, 4])
table.scale(1.2, 1.2)

# Apply Colors & Formatting
for (i, j), cell in table.get_celld().items():
    if i == 0:  # Header row
        cell.set_facecolor("#2980B9")  # Blue background
        cell.set_text_props(weight='bold', color='white', ha='center')
        cell.set_edgecolor('black')
    elif i == len(trial_balance_df) - 1:  # Total row
        cell.set_facecolor("#E74C3C")  # Red background
        cell.set_text_props(weight='bold', color='white', ha='center')
    else:
        cell.set_text_props(ha='center', fontsize=10, color='black')

# Hide axes
ax.axis('tight')
ax.axis('off')

# Show the Matplotlib Table
plt.show()
```
